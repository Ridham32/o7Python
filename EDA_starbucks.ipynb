```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("starbucks-menu-nutrition-drinks.csv")
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df["Calories"] = df["Calories"].astype(float)
```

```python
df.info()
```

```python
convFloat=["Fat (g)","Carb. (g)","Fiber (g)","Protein","Sodium"]
df[convFloat] = df[convFloat].apply(pd.to_numeric)
```

```python
df.info()
```

```python
df.rename(columns={"Unnamed: 0":"Drinks"},inplace=True)
```

```python
df.nlargest(2,"Calories")[["Drinks","Calories"]]
```

```python
df.nsmallest(10,"Calories")[["Drinks","Calories"]]
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="Carb. (g)",y="Drinks",data=df.sort_values("Carb. (g)",ascending=False).head(10))
plt.title("Top 10 Carb Drinks")
plt.show()
```
