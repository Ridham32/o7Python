```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =sns.load_dataset('titanic')
df
```

```python
sns.countplot(df,x='embarked')
```

```python
sns.countplot(df,x='survived',hue='class',width=0.3)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival count by class')
```
